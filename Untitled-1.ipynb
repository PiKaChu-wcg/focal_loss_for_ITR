```python
from model import *
from main import MyModel
from loss import *
from datasets import *
from tqdm import tqdm
import torch
torch.cuda.set_device(4)
import numpy as np
n=1000
```

```python

for name in ['focalinfonce_3050_128', 'infonce_128', "triplet_128"]:
    dist='../res/'+name+'.npy'
    opts = dict(
        vocab_path='./vocab/',
        dim_image=1280,
        dim=512,
        dim_word=256,
        margin=0.2,
        epochs=30,
        batch_size=64,
        lrate=0.001,
        gpus=[0,1,2,3],
        detach=False
    )
    opts['loss']=InfoNCE(0.1,learnable=True)
    model_dict=MyModel(opts)
    model_dict.load_state_dict(torch.load(f'/data/wuchangguang/school/VSE_Pytorch/lightning_logs/{name}/checkpoints/last.ckpt')['state_dict'],strict=False)
    model_dict.cuda()
    test=Flickr30K(split='test')
    test_loader, test_dataset = get_loader(
        test, model_dict.tokenizer, batch_size=opts['batch_size']*2, shuffle=False)
    pbar=tqdm(test_loader,total=len(test)/opts['batch_size']*2)
    IM=[]
    TXT=[]
    model_dict.eval()
    with torch.no_grad():
        for (txt,mask),im in pbar:
            txt=model_dict.model.forward_sens(txt.cuda(),mask.cuda())
            im=model_dict.model.forward_imgs(im.cuda())
            IM.append(im.cpu())
            TXT.append(txt.cpu())
    TXT = torch.concat(TXT, dim=0)
    IM = torch.concat(IM, dim=0)
    dis=IM.mm(TXT.T)
    np.save(dist,dis.numpy())
```
